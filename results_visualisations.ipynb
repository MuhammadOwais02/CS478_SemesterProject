```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("kruskalresults.csv")
```

```python
df.columns
```

```python
grouped = df.groupby(' Density')
```

```python
# Plot the line graphs
fig, ax = plt.subplots()

for name, group in grouped:
    ax.plot(group['Number of Nodes'], group[' Execution Time (Seconds)'], marker='o', label=f'Density {name}')

# Set labels and title
ax.set_xlabel('Number of Nodes')
ax.set_ylabel(' Execution Time (Seconds)')
ax.set_title('(Kruskal)- Execution Time vs Number of Nodes for Different Densities')

# Add legend
ax.legend()
fig.savefig('./plots/kruskal_result_visualisations/kruskal_line_graph.png')
# Show the plot
plt.show()
```
